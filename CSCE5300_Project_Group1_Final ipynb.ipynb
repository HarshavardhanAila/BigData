# PRE-PROCESSING

```python
#Pre processing
from pyspark.sql.functions import mean, col
from pyspark.ml import Pipeline
from pyspark.ml.feature import StringIndexer, OneHotEncoder, VectorAssembler

# Get the list of column names from the DataFrame
columns_for_processing = df.columns

# Define a list to hold all the stages for the pipeline
stages = []

# Handle missing values and perform encoding
for column in columns_for_processing:
    # Skip columns with non-string or non-numeric data types
    if df.schema[column].dataType not in ['string', 'integer', 'double']:
        continue

    if df.schema[column].dataType in ['integer', 'double']:
        mean_value = df.select(mean(col(column))).collect()[0][0]
        df = df.fillna(mean_value, subset=[column])

    if df.schema[column].dataType == 'string':
        indexer = StringIndexer(inputCol=column, outputCol=f"{column}_index")
        encoder = OneHotEncoder(inputCol=f"{column}_index", outputCol=f"{column}_encoded")
        stages += [indexer, encoder]

# Assemble all encoded columns into a single 'features' column
assembler = VectorAssembler(inputCols=[f"{column}_encoded" for column in columns_for_processing if df.schema[column].dataType == 'string'],
                            outputCol="features")
stages += [assembler]

# Create a pipeline with all stages
pipeline = Pipeline(stages=stages)

# Remove duplicate rows
df_no_duplicates = df.dropDuplicates()

# Fit and transform the data
transformed_data = pipeline.fit(df_no_duplicates).transform(df_no_duplicates)
df = transformed_data

# Show the transformed data
df.show()
df.count()

```

# DATA ANALYTICS (MACHINE LEARNING)

### **LINEAR REGRESSION**

```python
#creating new dataframe for linear regression
df1 = df.select("SMOKING", "ALCOHOL CONSUMING", "LUNG_CANCER")
df1.show(5)
```

```python
#changing values to ones and zeros (for yes and no)
from pyspark.sql.functions import when
df1 = df1.withColumn("SMOKING", when(df1.SMOKING == "2","1").when(df1.SMOKING == "1","0").otherwise(df1.SMOKING))
df1 = df1.withColumn("ALCOHOL CONSUMING", when(df1['ALCOHOL CONSUMING'] == "2","1").when(df1['ALCOHOL CONSUMING'] == "1","0").otherwise(df1['ALCOHOL CONSUMING']))
df1 = df1.withColumn("LUNG_CANCER", when(df1.LUNG_CANCER == "2","0").when(df1.LUNG_CANCER == "1","1").otherwise(df1.LUNG_CANCER))
```

```python
df1.show(5)
```

```python
#people who smoke & have cancer
smokingandlungcancer = df1.filter((df1.SMOKING == 1) & (df1['ALCOHOL CONSUMING'] == 0) & (df1.LUNG_CANCER == 1)).count()
#people who drink alcohol & have cancer
alcoholandlungcancer = df1.filter((df1.SMOKING == 0) & (df1['ALCOHOL CONSUMING'] == 1) & (df1.LUNG_CANCER == 1)).count()
#people who smoke & drink alcohol & have cancer
alcoholsmokingcancer = df1.filter((df1['ALCOHOL CONSUMING'] == 1) & (df1.LUNG_CANCER == 1) & (df1.SMOKING == 1)).count()
#people who don't smoke & don't drink alcohol & have cancer
lungcancer = df1.filter((df1['ALCOHOL CONSUMING'] == 0) & (df1.LUNG_CANCER == 0) & (df1.SMOKING == 1)).count()

import matplotlib.pyplot as plt
import numpy as np

#creating pie chart
arr = np.array([smokingandlungcancer, alcoholandlungcancer, alcoholsmokingcancer, lungcancer])
labels = ["Smoking & Lung Cancer", "Alcohol & Lung Cancer", "Smoking & Alcohol & Lung Cancer", "No Smoking & No Alcohol & Lung Cancer"]

#plotting pie chart
plt.pie(arr, labels = labels)
plt.legend(arr)
plt.show()
```

```python
#changing columns from string to integer
from pyspark.sql.types import IntegerType
for i in range(0, len(df1.columns)):
  df1 = df1.withColumn(df1.columns[i], df1[df1.columns[i]].cast(IntegerType()))
```

```python
# importing pypsark libraries
from pyspark.ml.tuning import ParamGridBuilder, CrossValidator
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.ml.regression import LinearRegression

#feature factorization
vector_assembler = VectorAssembler(inputCols=['SMOKING', 'ALCOHOL CONSUMING'], outputCol='features')
data = vector_assembler.transform(df1)

#splitting into training and testing data
final_data = data.select("features", "LUNG_CANCER")
train_data, test_data = final_data.randomSplit([0.8, 0.2], seed=42)
```

```python
train_data.describe().show()
```

```python
test_data.describe().show()
```

```python
#linear regression
Linear = LinearRegression(featuresCol="features", labelCol="LUNG_CANCER")
#fitting & transforming the data
lr_model = Linear.fit(train_data)
predictions = lr_model.transform(test_data)

#evaluating linear regression model
evaluator = RegressionEvaluator(labelCol="LUNG_CANCER", metricName="rmse")
rmse = evaluator.evaluate(predictions)
evaluator1 = RegressionEvaluator(labelCol="LUNG_CANCER", metricName="r2")
r2 = evaluator1.evaluate(predictions)
```

```python
#printing metrics
print("Root mean squared error: ", rmse)
print("R squared: ", r2)
```

```python
#printing coefficients & intercept used by model
coefficients = lr_model.coefficients
intercept = lr_model.intercept
print("Coefficients: ", coefficients)
print("Intercept: ", intercept)
```

The [absolute value] of the coefficient shows how much it affected/impacted the outcome, so the higher the absolute value of a coefficient the more it factors into the prediction for the outcome variable.


*   For smoking, the absolute value of the coefficient is 0.00296
*   For alcohol consuming, the absolute value of the coefficient is 0.00146

So, both, smoking and alcohol consuming, don't necessarily have a big impact or affect in a person being diagnosed with lung cancer.

### **DECISION TREE**

```python
from pyspark.sql import SparkSession
from pyspark.ml.feature import VectorAssembler

# Create a Spark session
spark = SparkSession.builder.appName("FeatureExtraction").getOrCreate()

# Assuming 'df' is your DataFrame with the columns you mentioned
feature_cols = ['GENDER', 'AGE', 'SMOKING', 'YELLOW_FINGERS', 'ANXIETY', 'PEER_PRESSURE', 'CHRONIC DISEASE',
                'WHEEZING', 'ALCOHOL CONSUMING', 'COUGHING', 'SHORTNESS OF BREATH',
                'SWALLOWING DIFFICULTY', 'CHEST PAIN']

# Create VectorAssembler to assemble selected columns into a single feature vector
vector_assembler = VectorAssembler(inputCols=feature_cols, outputCol='Lung_cancer_features')

# Apply VectorAssembler to transform the DataFrame
transformed_df = vector_assembler.transform(df)

# Show the resulting DataFrame with the feature vector
transformed_df.show()

```

```python
transformed_df.show()
```

```python
# importing pypsark and decision tree classifier
from pyspark.ml.tuning import ParamGridBuilder, CrossValidator
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
from pyspark.ml.classification import DecisionTreeClassifier
from pyspark.ml.feature import VectorAssembler

# Decision Tree Classifier
dt_classifier = DecisionTreeClassifier(featuresCol='Lung_cancer_features', labelCol='LUNG_CANCER')

# ParamGrid for hyperparameter tuning
param_grid = (ParamGridBuilder()
              .addGrid(dt_classifier.maxDepth, [5, 9, 11])
              .addGrid(dt_classifier.maxBins, [14, 24])
              .addGrid(dt_classifier.impurity, ["gini", "entropy"])
              .build())

# Evaluator for model performance (including accuracy)
evaluator = MulticlassClassificationEvaluator(labelCol="LUNG_CANCER", predictionCol="prediction", metricName="f1")
accuracy_evaluator = MulticlassClassificationEvaluator(labelCol="LUNG_CANCER", predictionCol="prediction", metricName="accuracy")

# CrossValidator for model tuning
crossval = CrossValidator(estimator=dt_classifier,
                          estimatorParamMaps=param_grid,
                          evaluator=evaluator,
                          numFolds=5)

# Fit the CrossValidator to the df data to tune the model
cv_model = crossval.fit(transformed_df)

# Get the best model and its hyperparameters
best_dt_model = cv_model.bestModel
best_maxDepth = best_dt_model.getOrDefault("maxDepth")
best_maxBins = best_dt_model.getOrDefault("maxBins")
best_impurity = best_dt_model.getOrDefault("impurity")

# Make predictions on the df data using the best model
best_predictions = best_dt_model.transform(transformed_df)

# Evaluate the model performance (F1 Score and Accuracy)
best_f1_score = evaluator.evaluate(best_predictions)
best_accuracy = accuracy_evaluator.evaluate(best_predictions)

# Print the best parameters, F1 score, and Accuracy
print(f"Best maxDepth: {best_maxDepth}")
print(f"Best maxBins: {best_maxBins}")
print(f"Best impurity: {best_impurity}")
print(f"Best F1 Score: {best_f1_score}")
print(f"Accuracy: {best_accuracy}")
```
